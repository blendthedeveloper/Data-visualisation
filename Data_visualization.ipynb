```python
# visualize.py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df=pd.read_csv(r"C:\Users\abhin\OneDrive\Documents\ABHIN\Projects\laptop_data.csv")
```

--Data Cleaning--

```python
# Remove 'GB' from RAM and convert to int
df['Ram'] = df['Ram'].str.replace('GB', '').astype(int)
```

```python
# Extract storage type (SSD/HDD/Flash)
df['Storage_Type'] = df['Memory'].apply(lambda x: 'SSD' if 'SSD' in x else ('HDD' if 'HDD' in x else 'Flash'))
```

```python
# Convert Price to numeric
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

--Visualization--

```python
# 1: Price Distribution

plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=30, kde=True, color='skyblue')
plt.title("Laptop Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.savefig(r"C:\Users\abhin\OneDrive\Documents\ABHIN\Projects\price_distribution.png")
plt.show()
```

```python
#2: Average Price by Brand

plt.figure(figsize=(12,6))
sns.barplot(data=df, x="Company", y="Price", hue="Storage_Type", ci=None, palette="viridis")
plt.title("Average Laptop Price by Brand (Grouped by Storage Type)")
plt.xticks(rotation=45)
plt.ylabel("Average Price")
plt.savefig(r"C:\Users\abhin\OneDrive\Documents\ABHIN\Projects\avg_price_by_brand_hue.png")
plt.show()
```

```python
#3: RAM vs Price

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Ram', y='Price', hue='Company', alpha=0.7)
plt.title("RAM vs Price by Brand")
plt.xlabel("RAM (GB)")
plt.ylabel("Price")
plt.savefig(r"C:\Users\abhin\OneDrive\Documents\ABHIN\Projects\ram_vs_price.png")
plt.show()
```

```python
#4: Storage Type vs Price

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Storage_Type", y="Price", palette="Set2")
plt.title("Price Comparison by Storage Type (Grouped by Company)")
plt.savefig(r"C:\Users\abhin\OneDrive\Documents\ABHIN\Projects\storage_vs_price_.png")
plt.show()
```

```python
#5: Correlation Heatmap

plt.figure(figsize=(8,6))
sns.heatmap(df[['Ram','Inches','Weight','Price']].corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.savefig(r"C:\Users\abhin\OneDrive\Documents\ABHIN\Projects\correlation_heatmap.png")
plt.show()
```
